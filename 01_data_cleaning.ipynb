#### __orders__:
- __order_id__: unique identifier of order
- __order_status__: delivered, canceled, etc
- __order_purchase_timestamp__: when order was placed
- __order_approved_at__: when order was approved
- __order_delivered_carrier_date__: date when the order is handed over to the carrier for shipment
- __order_delivered_customer_date__: actual date the order is delivered to the customer
- __order_estimated_delivery_date__: predicted date by which the customer is expected to receive the order
#### __order_items__:
- __order_id__: Unique identifier linking all items in a single order
- __order_item_id__: Sequential number identifying each item within the same order (e.g., first item, second item)
- __product_id__: Unique identifier of the product sold
- __seller_id__: Unique identifier of the seller fulfilling the item
- __shipping_limit_date__: Deadline by which the seller must hand over the item to the logistics partner
- __price__: Final sale price of the individual item
- __freight_value__: Shipping cost attributed to the item (shared across items if multiple in one order)

#### __products__:
- __product_id__: Unique identifier for the product
- __product_category_name__: Category to which the product belongs (e.g., electronics, furniture)
- __product_name_lenght__: Number of characters in the product's name
- __product_description_lenght__: Number of characters in the product's description
- __product_photos_qty__: Number of images associated with the product
- __product_weight_g__: Weight of the product in grams
- __product_length_cm__: Length of the product in centimeters
- __product_height_cm__: Height of the product in centimeters
- __product_width_cm__: Width of the product in centimeters 

#### __products_category__:
- __product_category_name__: Category name in Portuguese (original language of the dataset)
- __product_category_name_english__: English translation of the product category name


#### __customers__:
- __customer_id__: Identifier for a specific order transaction; the same customer can have multiple customer_ids across different orders
- __customer_unique_id__: Unique identifier for a customer across all orders; used to identify and track individual customers
- __customer_zip_code_prefix__: First 5 digits of the customer’s postal code
- __customer_city__: City where the customer is located
- __customer_state__: State where the customer is located

#### __sellers__:
- __seller_id__: Unique identifier for the seller
- __seller_zip_code_prefix__: First 5 digits of the seller’s postal code
- __seller_city__: City where the seller is located
- __seller_state__: State where the seller is located

#### __payments__:
- __order_id__: Unique identifier for the order
- __payment_sequential__: Sequence number for multiple payment methods used in a single order (e.g., 1st, 2nd payment)
- __payment_type__: Method of payment (e.g., credit card, debit card)
- __payment_installments__: Number of installments selected for the payment (e.g., 1 = full payment, 3 = split into 3 payments)
- __payment_value__: Transaction amount for that specific payment

#### __reviews__:
- __review_id__: Unique identifier for the customer review
- __order_id__: Identifier linking the review to a specific order
- __review_score__: Rating given by the customer, from 1 (lowest) to 5 (highest)
- __review_comment_title__: Short headline summarizing the review
- __review_comment_message__: Detailed feedback or comment provided by the customer
- __review_creation_date__: Timestamp when the review was written by the customer
- __review_answer_timestamp__: Timestamp when the seller or marketplace responded to the review

#### __geolocation__:
- __geolocation_zip_code_prefix__: First few digits of the postal code used for geographic identification
- __geolocation_lat__: Latitude coordinate of the zip code area
- __geolocation_lng__: Longitude coordinate of the zip code area
- __geolocation_city__: City corresponding to the zip code
- __geolocation_state__: State corresponding to the zip code

```python
order_details_agg = ( 
    order_items.groupby('order_id')
        .agg( order_value = ('price','sum'),
              total_freight_value = ('freight_value','sum'),
              item_count=('order_item_id','count')
            ).reset_index())
```

```python
payment_agg = (
    payments.groupby('order_id')
        .agg( total_payment_value =('payment_value','sum'),
              payment_installments=('payment_installments','max'),
              payment_type=('payment_type',lambda x:x.mode()[0])
            ).reset_index()
)
```

```python
orders['delivery_time_days']=(orders['order_delivered_customer_date']-orders['order_purchase_timestamp']).dt.days
orders['delivery_delay_days']=(orders['order_delivered_customer_date']-orders['order_estimated_delivery_date']).dt.days
orders['is_delayed']=orders['delivery_delay_days']>0
```

```python
order_level = (orders
    .merge(order_details_agg,on='order_id',how='left')
    .merge(payment_agg,on='order_id',how='left')
    .merge(reviews_clean,on='order_id',how='left')          
    )
```

```python
order_level.info()
```

```python
import pandas as pd

# join customers → orders
cust_orders = orders.merge(
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)

# join order_level
cust_orders = cust_orders.merge(
    order_level[['order_id', 'order_value']],
    on='order_id',
    how='left'
)

# join reviews
cust_orders = cust_orders.merge(
    reviews_agg,
    on='order_id',
    how='left'
)

```

```python
print(cust_orders.columns)   
```

```python
customer_level = (
    cust_orders
    .groupby('customer_unique_id')
    .agg(
        first_purchase_date=('order_purchase_timestamp', 'min'),
        last_purchase_date=('order_purchase_timestamp', 'max'),
        total_orders=('order_id', 'nunique'),
        total_revenue=('order_value', 'sum'),
        avg_delivery_delay=('delivery_delay_days', 'mean')
    )
    .reset_index()
)

order_level = order_level.merge(
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)
```

```python
customer_level
```

```python
seller_level = (
    order_items
    .groupby('seller_id')
    .agg(
        total_orders=('order_id','nunique'),
        total_revenue=('price','sum')
        ).reset_index()
    )
```

```python
order_level = order_level.copy()

order_level = order_level.drop_duplicates(subset=['order_id'])

seller_order = (
    order_items
    .groupby(['order_id', 'seller_id'], as_index=False)
    .agg(
        item_count=('order_item_id', 'count'),
        seller_order_revenue=('price', 'sum'),
        seller_freight_revenue=('freight_value', 'sum')
    )
)

seller_order['seller_order_revenue'] = (
    seller_order['seller_order_revenue'] +
    seller_order['seller_freight_revenue']
)


seller_order = seller_order.merge(
    order_level[['order_id', 'is_delayed', 'delivery_delay_days', 'review_score']],
    on='order_id',
    how='left'
)

seller_level = (
    seller_order
    .groupby('seller_id', as_index=False)
    .agg(
        total_orders=('order_id', 'nunique'),
        total_revenue=('seller_order_revenue', 'sum'),
        delayed_orders=('is_delayed', lambda x: (x == True).sum()),
        avg_delivery_delay=('delivery_delay_days', 'mean'),
        avg_review_score=('review_score', 'mean'),
        one_star_pct=('review_score', lambda x: (x == 1).mean() * 100)
    )
)

seller_level['delay_pct'] = (
    seller_level['delayed_orders'] /
    seller_level['total_orders'] * 100
)
```

```python
# - Which product categories have highest delays?
# - Which products drive revenue vs. problems?
# - Category-specific seller performance
# product_level will help analyse above business problems

product_level = order_items.merge(
    products[['product_id', 'product_category_name']], 
    on='product_id'
).merge(
    orders[['order_id','review_score', 'is_delayed', 'delivery_delay_days']], 
    on='order_id'
).groupby('product_category_name').agg({
    'order_id': 'nunique',                    # total_orders
    'price': 'sum',                            # total_revenue
    'freight_value': 'sum',                    # total_freight
    'is_delayed': 'sum',                       # delayed_orders
    'delivery_delay_days': 'mean',             # avg_delay
    'review_score': 'mean'                     # avg_rating
}).reset_index()

product_level.columns = [
    'product_category', 'total_orders', 'total_revenue', 
    'total_freight', 'delayed_orders', 'avg_delivery_delay', 'avg_review_score'
]

product_level['delay_pct'] = (
    product_level['delayed_orders'] / product_level['total_orders'] * 100
)

product_level['revenue_per_order'] = (
    product_level['total_revenue'] / product_level['total_orders']
)
```
